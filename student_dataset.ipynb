```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['gender'].value_counts().plot.bar()
```

```python
df['writing score'].value_counts(normalize = True).plot.bar(figsize=(20,10),title='level')
```

```python
df['parental level of education'].value_counts(normalize = True).plot.bar(figsize=(15,10),title='Education level')
```

```python
counts_lunch = df['lunch'].value_counts()
counts_lunch = counts_lunch.reset_index()
counts_lunch.columns = ["lunch","Count"]
print(counts_lunch)
```

```python
pt = sns.catplot(y="Count",
            x="lunch", data=counts_lunch,
           palette="colorblind",height=5, aspect=2, kind="bar");
plt.style.use("ggplot")
plt.title("Figure 1 : \n\n lunch Count Plot",
          fontsize=25);
```

```python
parented = df['parental level of education'].value_counts()
parented = parented.reset_index()
parented.columns = ["parental level of education","Count"]
print(parented)
```

```python
pt = sns.catplot(y="Count",
            x="parental level of education", data=parented,
           palette="bright",height=5, aspect=2, kind="bar");
plt.style.use("ggplot")
plt.title("Figure 1 : \n\n Parental level of education Plot",
          fontsize=25);
```

```python
sns.lmplot(x="reading score", y="writing score",data=df,markers="v",
            scatter_kws={"color": "blue"},
         line_kws={"linewidth":3,"color":"red"},aspect=2);
plt.title("Figure : \n\n\n Reading score vs Writing score");
```

```python
sns.lmplot(x="math score", y="writing score",data=df,markers="o",
           order=3, ci=None, scatter_kws={"color": "green"},
         line_kws={"linewidth":2,"color":"blue"},aspect=2);

plt.title("Figure : \n\n\n Math score vs Writing score");
```

```python
sns.jointplot(x="math score", y="reading score",data=df,
            
            kind="kde",height=10,palette="colorblind");
plt.title("Figure : \n\n\n Math Score vs Reading score", fontsize=15);
```
